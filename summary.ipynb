### 연산자

```python
# 관계연산자:
#  > , >=, <, <=, ==, !=

a = 10
b = 20
print(a > b)  # False
print(a >= b)  # False
print(a < b)  # True
print(a <= b)  # True
print(a == b)  # False
print(a != b)  # True

# 논리연산자: 결과 - True, False
#  and, or, not

print(True and False)  # False
print(True or False)  # True
print(not True)  # False

# 비트연산자:
#  &, |, ^, ~, <<, >> (쉬프트연산자)

print(bin(10 & 12))  # 0b1000
print(bin(10 | 12))  # 0b1110
print(bin(10 ^ 12))  # 0b110
print(bin(~10))  # -0b1011
print(bin(10 << 1))  # 0b10100
print(bin(10 >> 1))  # 0b101

# 증감연산자(++, --): 지원하지 않음
#  i+=1, i-=1 등 대신 사용

i = 10
i += 1
print(i)  # 11
i -= 1
print(i)  # 10

# 사칙연산자:
#  +, -, *, /, %(나머지)

print(10 + 20)  # 30
print(10 - 20)  # -10
print(10 * 20)  # 200
print(10 / 20)  # 0.5
print(10 % 20)  # 10

# bool 로 만들기: bool(1) 

print(bool(1))  # True
print(bool(0))  # False
print(bool(""))  # False
print(bool("Hello"))  # True

# 객체비교: is, is not

a = [1, 2, 3]
b = [1, 2, 3]
print(a is b)  # False
print(a is not b)  # True

# 형변환:
#  a = int("10") # 정수형이 됨
#  a = float("10.5") # 실수형이 됨
#  a = str(10) # 문자열이 됨

a = int("10")
print(type(a))  # <class 'int'>
a = float("10.5")
print(type(a))  # <class 'float'>
a = str(10)
print(type(a))  # <class 'str'>

# 복합연산자:
#  +=, -=, *=, /=, %=

a = 10
a += 5
print(a)  # 15
a -= 5
print(a)  # 10
a *= 5
print(a)  # 50
a /= 5
print(a)  # 10.0
a %= 5
print(a)  # 0.0

# 조건연산자(3항연산자):
#  (a > b)? a: b

a = 10
b = 20
result = a if a > b else b
print(result)  # 20
```

### 딕셔너리 - { }로 생성. 키/값쌍 자료형

```python
d = {'a': 10, 'b': 20}  # 기본 형식, 키가 중복되면 나중것만 유효 
x = {}
x = dict()  # 빈딕셔너리 만들기 

# dict 함수는 리스트, 튜플, 딕셔너리로 만들 때 사용.
d1 = dict(키1=10, 키2=20)
print(d1)  # {'키1': 10, '키2': 20}

d2 = dict(zip(['키1','키2'],[10,20]))
print(d2)  # {'키1': 10, '키2': 20}

d3 = dict([('키1',10), ('키2',20)])
print(d3)  # {'키1': 10, '키2': 20}

d4 = dict({'키1':10, '키2':20})
print(d4)  # {'키1': 10, '키2': 20}

d = {'a':10, 'b':20}
d['a'] = 30  # 키값으로 접근하고, 할당 
print(d)  # {'a': 30, 'b': 20}

print('a' in d)  # True : 키가 있는지 확인 
print(len(d))  # 2 :항목 갯수
```

### 딕셔너리 다루기 - 메서드, 출력

```python
# 키/값 쌍 추가 
x = {'a':10, 'b':20}

# setdefault() 메서드 사용
x.setdefault('c', 30)  # 키 'c'가 없으면 {'c': 30}을 추가하고 30을 반환, 있으면 해당 값을 반환
print(x)  # {'a': 10, 'b': 20, 'c': 30}

# update() 메서드 사용
x.update(a=5, d=40)  # 업데이트는 ' " 따옴표 빼고 사용. 키값이 없으면 추가됨.
print(x)  # {'a': 5, 'b': 20, 'c': 30, 'd': 40}

x.update({1:'One', 2:'Two'})  # 딕셔너리도 가능 
print(x)  # {'a': 5, 'b': 20, 'c': 30, 'd': 40, 1: 'One', 2: 'Two'}

x.update([[3,'Three'],[4,'Four']])  # 리스트도 가능
print(x)  # {'a': 5, 'b': 20, 'c': 30, 'd': 40, 1: 'One', 2: 'Two', 3: 'Three', 4: 'Four'}

# pop() 메서드 사용
value = x.pop('a')  # 키에 해당하는 항목을 삭제하고 값을 리턴
print(value)  # 5
print(x)  # {'b': 20, 'c': 30, 'd': 40, 1: 'One', 2: 'Two', 3: 'Three', 4: 'Four'}

value = x.pop('e', 0)  # 키 'e'가 있으면 삭제후 해당값을 없으면 지정한값(0) 을 리턴.
print(value)  # 0

# popitem() 메서드 사용
item = x.popitem()  # 마지막 항목을 삭제하고 튜플로 리턴.
print(item)  # (4, 'Four')
print(x)  # {'b': 20, 'c': 30, 'd': 40, 1: 'One', 2: 'Two', 3: 'Three'}

# clear() 메서드 사용
x.clear()  # 모두삭제 
print(x)  # {}

x = {'a': 10, 'b': 20}

# get() 메서드 사용  
value = x.get('a', 0)  # 키의 값 가져오기. 없으면 기본값(0) 반환.
print(value)  # 10

value = x.get('c', 0)  # 키의 값 가져오기. 없으면 기본값(0) 반환.
print(value)  # 0

# items() 메서드 사용
items = x.items()  # 키-값 쌍 모두 가져옴
print(items)  # dict_items([('a', 10), ('b', 20)])

# keys() 메서드 사용
keys = x.keys()  # 키값을 모두 가져옴 
print(keys)  # dict_keys(['a', 'b'])

# values() 메서드 사용
values = x.values()  # 값을 모두 가져옴.
print(values)  # dict_values([10, 20])

# fromkeys() 메서드 사용
ks = ['a', 'b']
x = dict.fromkeys(ks)  # 값이 None 인 딕셔너리를 생성
print(x)  # {'a': None, 'b': None}

x = dict.fromkeys(ks, 100)  # 값이 100 인 딕셔너리 생성
print(x)  # {'a': 100, 'b': 100}

# defaultdict 사용
from collections import defaultdict

x = defaultdict(int)  # 키값이 없을 경우 디폴트값 처리함수 int() 는 0을 리턴하므로 0이 리턴됨.
x['a'] = 10
x['b'] = 20
print(x['c'])  # 0

x = defaultdict(lambda: 'Default')  # 람다표현식으로, 디폴트로 'Default' 리턴
print(x['d'])  # 'Default'

# 키, 값 모두 출력
for key, value in x.items():
   print(key, value)

# 키값 가져와서 루프
for key in x.keys():
   print(key)

# 값을 가져와서 루프  
for value in x.values():
   print(value)

# 딕셔너리 컴프리헨션
x = {key: value for key, value in dict.fromkeys(ks).items()}
print(x)  # {'a': None, 'b': None}

# 얕은 복사
a = {'a': 10, 'b': 20}
b = a.copy()
print(b)  # {'a': 10, 'b': 20}

# 깊은 복사
import copy

a = {'a': {'x': 1}, 'b': {'y': 2}}
b = copy.deepcopy(a)
print(b)  # {'a': {'x': 1}, 'b': {'y': 2}}
```
